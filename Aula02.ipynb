### Criar uma função para plotar o boxplot do seaborn

```
def sns_boxplot(data, x, y, title="", order=None, hue=None):
  fig, ax = plt.subplots(figsize=(10, 6))
  sns.boxplot(ax=ax, x=x, y=y, data=data, order=order, hue=hue)
  ax.set_title(title)
  fig.savefig(title+'.png')
```

```
sns_boxplot(data=dados, x="Q006", y="NU_NOTA_TOTAL", title="Boxplot das notas por classe social com os alunos faltosos", order=renda_ordenada )
```

```
sns_boxplot(data=dados_sem_notas_zero, x="Q006", y = "NU_NOTA_TOTAL", order = renda_ordenada, title="Boxplot das notas por classe social sem os alunos faltosos")
```

### Verificar se quem zerou a prova foi eliminado ou não estava presente

```
presenca = ["TP_PRESENCA_CN", "TP_PRESENCA_CH", "TP_PRESENCA_LC", "TP_PRESENCA_MT", "NU_NOTA_CN", "NU_NOTA_CH", "NU_NOTA_LC", "NU_NOTA_MT"]
dados_fez_prova_zerou = dados[presenca].query("NU_NOTA_CH==0 or NU_NOTA_CN==0 or NU_NOTA_LC==0 or NU_NOTA_MT==0").groupby(["TP_PRESENCA_CN", "TP_PRESENCA_CH", "TP_PRESENCA_LC", "TP_PRESENCA_MT"])

dados_fez_prova_zerou.head()
```

###  Verificar a proporção dos participantes de rendas mais altas e mais baixas como treineiro e não treineiro.

```
sns_boxplot(data=dados_sem_notas_zero, x="Q006", y = "NU_NOTA_TOTAL",
            order = renda_ordenada,
            title="Boxplot das notas por classe social sem os alunos faltosos",
            hue="IN_TREINEIRO")
```

```
fig, ax = plt.subplots(nrows=2, figsize=(16,8))

sns.histplot(ax=ax[0], data=dados_sem_notas_zero[(dados_sem_notas_zero.IN_TREINEIRO==0)].sort_values("Q006"),
             x="Q006",
             stat="probability")
sns.histplot(ax=ax[1], data=dados_sem_notas_zero[(dados_sem_notas_zero.IN_TREINEIRO==1)].sort_values("Q006"),
             x="Q006",
             stat="probability")
ax[0].set_xlabel("")
ax[0].set_title("Proporção dos participantes não treineiro de acordo com a renda")
ax[1].set_title("Proporção dos participantes treineiro de acordo com a renda")
fig.savefig("proporção dos participantes de rendas mais altas e mais baixas como treineiro e não treineiro.png")
```

```
dados_agrupados_por_Q006=(dados.groupby('Q006')['IN_TREINEIRO']
          .value_counts(normalize=True).rename('porcentagem').reset_index().sort_values('Q006'))

plt.figure(figsize=(16,8))
#Plotando gráfico
ax=sns.barplot(x='Q006', y='porcentagem', data=dados_agrupados_por_Q006, hue="IN_TREINEIRO", palette='RdBu')

#Colocando Titulo e Etiquetas nos eixos
ax.set_title('Participantes treineiros e não-treineiros por renda', fontsize=18)
ax.set_xlabel('renda', fontsize=14)

#Colocando as porcentagens em cima das barras
for p in ax.patches:
    val = '{:.2f}'.format(p.get_height())
    x = p.get_x()
    y = p.get_y() + p.get_height()
    ax.annotate(val, (x, y+.01), rotation=45)

plt.savefig("Participantes treineiros e não-treineiros por renda.png")
```
